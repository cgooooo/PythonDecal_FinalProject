```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.optimize as fit
```

```python
file = 'cluster_catalog.csv'
data = pd.DataFrame(pd.read_csv(file))

def distance():
    return (((data["z"]) * (2.998e5)) / 70)

def velocity():
    return (10**(data["logsigma"]))

def model(x, m, b):
    return m*x + b

fit_model = model
p0 = [1/70, 2.2]
pars, cov = fit.curve_fit(fit_model, distance(), velocity(), p0 = p0)



plt.figure(figsize = (20, 10))
plt.errorbar(distance(), velocity(), yerr = 10**data["e_logsigma"], fmt = "o")
plt.plot(distance(), model(distance(), pars[0], pars[1]))
plt.title('Hubble Constant')
plt.xlabel('Distance (Mpc)')
plt.ylabel('Velocity (Km/s)')
plt.show()

pars[0]
```

```python
#Travel time = diameter/ velocity dispersion 

import random
n = np.random.uniform(0, 20, 10)

veldisp = np.array(data['logsigma'])
veldisp = np.array(10**veldisp)

random_gal = np.random.randint(0,279,5)

time1 = (n*3.08e19/veldisp[0])*(3.15e-17)
time2 = (n*3.08e19/veldisp[1])*(3.15e-17)
time3 = (n*3.08e19/veldisp[2])*(3.15e-17)
time4 = (n*3.08e19/veldisp[3])*(3.15e-17)
time5 = (n*3.08e19/veldisp[4])*(3.15e-17)

plt.figure(figsize = (20, 10))
plt.scatter(n, time1, label = data["Name"][random_gal[0]])
plt.scatter(n, time2, label = data["Name"][random_gal[1]])
plt.scatter(n, time3, label = data["Name"][random_gal[2]])
plt.scatter(n, time4, label = data["Name"][random_gal[3]])
plt.scatter(n, time5, label = data["Name"][random_gal[4]])
plt.axhline(y=13.8, color='orange', label = "Hubble Time")
plt.title('Hubble Time')
plt.xlabel('Diameter (Mpc)')
plt.ylabel('Travel Time (Giga Years)')
plt.legend()
plt.show()
```
