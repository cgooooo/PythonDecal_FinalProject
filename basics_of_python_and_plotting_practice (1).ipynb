### Plotting with `Matplotlib`

```python
from matplotlib import pyplot as plt
import random
```

Define arrays for point coordinates

```python
# Define an array of x-values
x = [0, 1, 2, 3, 4, 5, 6]

# Define 3 empty arrays for y-values
y1, y2, y3 = [], [], []

# For every element in the x array,
#   add a random number between 0 and 10
#   to each y array
for i in range(len(x)):
    y1.append(random.randint(0, 10))
    y2.append(random.randint(0, 10))
    y3.append(random.randint(0, 10))
```

Plot a set of points

```python
# Make a figure for your plot
fig = plt.figure(figsize=(7, 4))

# Plot the series of points defined by
#   the x array and the 1st y array
#   with plt.plot
plt.plot(x, y1)

# Add axes labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# Show figure
plt.show()
```

```python
# Make a figure object
fig = plt.figure(figsize=(7, 4))

# Plot a circle at each point defined by the
#   x array and first y array using
#   plt.scatter
plt.scatter(x, y1)

# Label the axes
plt.xlabel('$x$')
plt.ylabel('$y$')

# Show the figure
plt.show()
```

Plot multiple lines on one plot

```python
fig = plt.figure(figsize=(7, 4))

# Plot multiple lines in a single figure and label each line
plt.plot(x, y1, label=r'Line 1 $\beta$')
plt.plot(x, y2, label='Line 2')
plt.plot(x, y3, label='Line 3')
plt.xlabel('$\gamma$')
plt.ylabel('$y$')

# Add a legend so that the lines are easy to identify
plt.legend(fontsize=14)
plt.show()
```

Plot multiple lines on separate subplots

```python
# Create a set of subplots (1 row, 3 columns)
fig, axs = plt.subplots(1, 3, figsize=(7, 4))

# Set first line in first subplot and label each axis
axs[0].plot(x, y1)
axs[0].set_xlabel('$x$')
axs[0].set_ylabel('$y$')

# Set second line in second subplot and label each axis
axs[1].plot(x, y2)
axs[1].set_xlabel('$x$')
axs[1].set_ylabel('$y$')

# Set third line in third subplot and label each axis
axs[2].plot(x, y3)
axs[2].set_xlabel('$x$')
axs[2].set_ylabel('$y$')
plt.show()
```

```python
# Same as above but make subplots share the y-axis to clean up the figure
fig, axs = plt.subplots(1, 3, figsize=(7, 4), sharey=True)

# Add each line to its own subplot
axs[0].plot(x, y1)
axs[1].plot(x, y2, color='purple') # You can also set color
axs[2].plot(x, y3, color='orange', marker='*') # And markers

# Add labels to the entire figure and scale up their fontsizes
axs[1].set_xlabel('$x$', fontsize=20)
axs[0].set_ylabel('$y$', fontsize=20)

# Add titles for each plot
axs[0].set_title('Left Plot')
axs[1].set_title('Middle Plot')
axs[2].set_title('Right Plot')

# Move the subplots closer together to reduce whitespace
plt.tight_layout()
plt.show()
```

`Numpy` gives you the flexibility to combine any number of functions to model equations you're interested in computing. To give an example of the power of `Numpy` for modeling complicated functions, we'll use some of `Numpy` math functions to model an arbitrary  function,

$$y\left(x\right) = \left[\sin\left(9 x^2 + {\pi \over 4}\right) + \cos(2 x)\right] e^{-x^2},$$

and then plot it using `matplotlib`.

```python
# Create an array of 1000 numbers equally spaced between 1 and 2
x = np.linspace(1, 2, 1000)

# Model the equation above using the array, x,
# and a combination of numpy functions
y = (np.sin(9 * np.square(x) + np.pi / 4) + np.cos(2 * x)) * np.exp(-x ** 2)
```

```python
# Plot the function above using the x and y arrays
plt.figure(figsize=(7, 5))
plt.plot(x, y)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.show()
```

### EXTRA: Numerical Integration and Fitting with `Scipy`

These functions likely won't be important for most of the class, but may be useful at the end of the year once we get to the cosmology unit.

```python
# Import the scipy library
from scipy import integrate
from scipy import optimize
```

$$y\left(x\right) = \frac{1}{2}x^2 + 3 x - 2$$

$$\int_{0}^{2} y\left(x\right) dx = \frac{1}{6} x^3 + \frac{3}{2}x^2 - 2x \bigg|_{0}^{2}$$

```python
# Create an array of values from 0 to 2
x_low, x_high = 0, 2
x = np.linspace(x_low, x_high, 100)
y = 0.5 * x ** 2 + 3 * x - 2

# Compute the true value of the integral
true_integral = (1 / 6 * x_high ** 3 + 3 / 2 * x_high ** 2 - 2 * x_high) - (1 / 6 * x_low ** 3 + 3 / 2 * x_low ** 2 - 2 * x_low)

# Compute the integral using scipy (trapezoid rule)
estimated_integral = integrate.trapz(y=y, x=x)

print(
    f"True integral value: {np.round(true_integral, 3)}\nEstimated value w/ scipy: {np.round(estimated_integral, 3)}"
)
```

```python
def polynomial(x, a, b, c):
    """
    """
    return a * x ** 2 + b * x + c
```

```python
# Generate an array of 50 random between 1 and 10
x = np.random.uniform(1, 10, size=50)
x = np.sort(x)

# Create a set a input numbers to be used in the polynomial function
parameters = {
    'a': 0.4,
    'b': -4,
    'c': np.pi / 4
}

# Generate a set of data from the function using the random numbers
# and input parameters
data = polynomial(x, **parameters)

# Add some noise to our simulated data
noise = np.random.normal(0, 0.05, size=50)
data += noise
```

```python
# Use scipy's build in curve_fit function to fit for the polynomial coefficients
# First input variable is the function to use as a model, the second are the values of
# the x-axis, and the third are the data. The function then returns the estimated fit
# parameters and a matrix describing the uncertainities in the estimates of the parameters
popt, pcov = optimize.curve_fit(polynomial, x, data)
```

```python
plt.figure(figsize=(7, 4))
plt.plot(x, polynomial(x, *popt), color='blue', label='Model')
plt.scatter(x, data, color='black', label='Data')
plt.xlabel('$x$', fontsize=20)
plt.ylabel('$y$', fontsize=20)
plt.legend(fontsize=14)
plt.show()
```

Let's now see how the fit parameters compare to the actual values. As we can see below, the fit values are quite close, but not exact due to the noise added to the data. This function ends up being extremely useful because of how simple `scipy` makes fitting a model to data.

```python
print ('Variable\t True Value\t Fit Value', )
for ki, (key, value) in enumerate(parameters.items()):
    print (f"{key}\t\t {np.round(value, 3)}\t\t {np.round(popt[ki], 3)}")
```
